# Maquinas de Vectores de Soporte (CVM)

### Ventajas
* Precision muy alta en comparacion a arboles de desicion y Regresion Logistica
* Funciona con el truco de Kernel (Genera un hiperplano 3D que mejor divida el conjunto de clases en una dimension superior)
* El hiperplano puede ser lineal, polinomico o rgb
* Es mas rapido que Naive Bayes
* Funciona bien cuando el margen de separacion es elevado

### Desventajas
* No son adecuadas para grandes volumenes de datos debido a su alto tiempo de formacion
* Funciona mal con clases superpuestas
* Sensible al tipo de Kernel o Nucleo

## KERNEL O NUCLEO

* Transforma de manera mas eficiente los datos a dimensiones superiores
* Tipos de Kernel: Lineal, polinomial, de base radial o rbf (Gaussiano)



```python
# iMPORTAMOS EL DATASET
from sklearn import datasets

dataset = datasets.load_breast_cancer()
```

```python
dataset.keys() #muestra las llaves para conocer el dataset
```

```python
x_data = dataset.data #Se presentan todos los datos el dataset. Mas de 1D
y_data = dataset.target
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2) #0.2 es la proporcion de division
```

## Definiendo el algoritmo

```python
from sklearn.svm import SVC
```

```python
algoritmo = SVC(kernel="linear")
```

```python
#ENTRENANDO EL MODELO
algoritmo.fit(x_train, y_train)
```

```python
#REALIZANDO PREDICCION
y_predict = algoritmo.predict(x_test)
```

## Verificando Los Resultados

```python
from sklearn.metrics import confusion_matrix
matriz = confusion_matrix(y_test, y_predict)
matriz

#Matriz de confusion
#print("[VN, FP]") #Verdaderos positivos, falsos negativos
#print("[FN, VP]") #falsos positivos, verdaderos negativos
```

## Precision del Modelo

```python
from sklearn.metrics import precision_score
precision_score(y_test, y_predict) # Arroja una buena prediccion
```
